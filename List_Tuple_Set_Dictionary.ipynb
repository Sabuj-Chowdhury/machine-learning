```python
#converting a list to tuple
new_list_1 =[1,3,2,4,"hello",False,18,"kuddos"]
print(new_list_1)
type(new_list_1)

new_converted_tuple = tuple(new_list_1)
print(new_converted_tuple)
type(new_converted_tuple)
```
